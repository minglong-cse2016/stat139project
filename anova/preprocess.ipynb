```python
import cPickle
f = open('../data/mnist.pkl', 'rb')
train_set, valid_set, test_set = cPickle.load(f)
f.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.reshape(train_set[0][1], (28,28))
plt.imshow(x[14:21, 14:21])
plt.show()
plt.matshow(x)
plt.show()
```

```python
anova_data = []

output = open('anova.csv', 'w')
row_num = 7
width = 28 / row_num
schema = ','.join(['r'+str(i) for i in range(row_num*row_num)])+',label\n'
output.write(schema)

for i in range(50000):
    cur_img = np.reshape(train_set[0][i], (28,28))
    label = train_set[1][i]
    patch_mean = np.zeros((row_num,row_num))
    for row in range(row_num):
        for column in range(row_num):
            patch = cur_img[row*width:(row+1)*width, column*width:(column+1)*width]
            patch_mean[row, column] = np.mean(patch[:])
    output.write(','.join([str(num) for num in patch_mean.reshape((1,row_num*row_num)).tolist()[0]]) + ',' + str(label)+'\n')
```
